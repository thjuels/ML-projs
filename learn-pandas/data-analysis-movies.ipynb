```python
test_str = "['2 hours 12 minutes']"
def convertTimeFromString(time_as_str):
    # time_as_str = "2 hours 12 minutes"
    if not isinstance(time_as_str, str):
        return time_as_str
    time_as_str = time_as_str.strip("[]'")
    # print(time_as_str)
    hours = 0
    minutes = 0
    if (time_as_str == "Not_Found"):
        return int(0) # return 0 for invalid ones
        
    split_str = time_as_str.split()
    
    # The i + 1 determines the time 'type' and i is value
    for i in range(len(split_str) - 1):
        if split_str[i + 1] == "hours":
            hours = int(split_str[i])
        elif split_str[i + 1] == "minutes":
            minutes = int(split_str[i])
    
    # We'll return minutes
    return hours * 60 + minutes

print(convertTimeFromString(test_str))
```

```python
df["Run_Time"] = df["Run_Time"].apply(convertTimeFromString)
print(type(df["Run_Time"].iloc[0]))
df.head()
```

```python
def convertReleaseDate(release_date):
    if not isinstance(release_date, str):
        return release_date

    release_date = release_date.strip("[]'")
    release_date_arr = release_date.split()

    if len(release_date_arr) < 3:
        return "Invalid Date Format"

    month_map = {
        "January": '01', "February": '02', "March": '03', "April": '04', "May": '05', "June": '06',
        "July": '07', "August": '08', "September": '09', "October": '10', "November": '11', "December": '12'
    }

    month = month_map.get(release_date_arr[0], "Invalid")
    day = release_date_arr[1].rstrip(',')
    day = day.zfill(2)
    year = release_date_arr[2]

    if month == "Invalid":
        return "Invalid Month Name" 

    return f"{day}/{month}/{year}"

print(convertReleaseDate(first_el_rd))
```

```python
df["Release_Date"] = df["Release_Date"].apply(convertReleaseDate)
# it seems Vote_average column is str, so will convert to int
df["Vote_Average"] = df["Vote_Average"].apply(lambda x: float(x))
df.head()
```

### Performing some data visualizations below

```python
def plotRatingDistribution():
    
    counts, bins, _ = plt.hist(df["Vote_Average"], bins=20, color="moccasin", alpha=0.6, edgecolor='black')
    plt.title('IMDb Ratings Distribution', fontsize=14)
    plt.xlabel('Rating', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    
    mean_rating = df["Vote_Average"].mean()
    std_rating = df["Vote_Average"].std()
    median_rating = df["Vote_Average"].median()

    # I want to add a curve that follows it also
    kde = ss.gaussian_kde(df["Vote_Average"].dropna(), bw_method=0.3)
    x = np.linspace(df["Vote_Average"].min(), df["Vote_Average"].max(), 100)
    y = kde(x) * (max(counts) / max(kde(x)))
    plt.plot(x, y, color="red", linewidth=2, label="Normal Fit")

    
    plt.axvline(x=mean_rating, color='blue', linestyle='--', 
                label=f'Mean: {mean_rating:.2f}')
    plt.axvline(x=median_rating, color='green', linestyle='-', 
                label=f'Median: {median_rating:.2f}')
    
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plotBudgetVsRevenue():    
    # breakeven line
    plt.figure()
    max_val = df["Budget"].max()
    colors = np.where(df["Revenue"] >= df["Budget"], "green", "red")
    plt.plot([0, max_val], [0, max_val], color="blue", linestyle="dashed", label="Breakeven Line")
    plt.scatter(df["Budget"], df["Revenue"], c=colors)

    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Budget (Log Scale)")
    plt.ylabel("Revenue (Log Scale)")
    plt.title("Budget vs Revenue (Log Scale)")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)  # Add grid
    plt.legend()
    plt.show()
```

```python
all_genres = []
# Need to convert df['Genres'] which is a string to an actual 
for genres in df['Genres'].apply(ast.literal_eval):
    all_genres.extend(genres)

top_genres = [genre for genre, count in Counter(all_genres).most_common(7)]

def plotRuntimeByGenre():

    plt.figure()
    genreVsRuntimeData = []
    for _, row in df.iterrows():
        genres = ast.literal_eval(row["Genres"])
        runtime = row["Run_Time"]
        for genre in genres:
            if genre in top_genres:
                genreVsRuntimeData.append({
                    'Genre': genre,
                    'Runtime': float(runtime)
                })
    data_df = pd.DataFrame(genreVsRuntimeData)
    sns.boxplot(data_df, x='Genre', y='Runtime', hue="Genre", whis=[0, 100], width=.6, palette="vlag")
    # To include observations
    sns.stripplot(data_df, x="Genre", y="Runtime", size=4, color=".3")
```

```python
def plotRatingByGenre():
    plt.figure()
    genreVsRuntimeData = []
    for _, row in df.iterrows():
        genres = ast.literal_eval(row["Genres"])
        rating = row["Vote_Average"]
        for genre in genres:
            if genre in top_genres:
                genreVsRuntimeData.append({
                    'Genre': genre,
                    'Rating': float(rating)
                })
    data_df = pd.DataFrame(genreVsRuntimeData)
    sns.boxplot(data_df, y='Genre', x='Rating', hue="Genre", whis=[0, 100], width=.6, palette="Accent")
    # To include observations
    sns.stripplot(data_df, y="Genre", x="Rating", size=4, color=".3")
```

```python
def plotReleaseTimeline():
    df["Release_Date"] = pd.to_datetime(df["Release_Date"], dayfirst=True)
    df["Release_M"] = df["Release_Date"].dt.to_period('M')
    # Average rating per month
    monthly_ratings = df.groupby('Release_M')['Vote_Average'].mean().reset_index()
    # # of movies per month
    monthly_counts = df.groupby('Release_M').size().reset_index(name='count')
    
    monthly_ratings['Release_M'] = monthly_ratings['Release_M'].astype(str)
    monthly_counts['Release_M'] = monthly_counts['Release_M'].astype(str)
    
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.plot(monthly_ratings['Release_M'], monthly_ratings['Vote_Average'], color="tab:blue")
    ax1.set_xlabel('Release Month')
    ax1.set_ylabel("Average Rating")
    
    ax2 = ax1.twinx()
    ax2.bar(monthly_counts["Release_M"], monthly_counts["count"], color="orange", alpha=0.6)
    ax1.set_xlabel('Number of Releases')
```

```python
def featurePairPlot():
    sns.pairplot(df)
```

```python
def analyzeData():
    plotRatingDistribution()
    plotRuntimeByGenre()
    plotRatingByGenre()
    plotBudgetVsRevenue()
    plotReleaseTimeline()
    featurePairPlot()

analyzeData()
```
