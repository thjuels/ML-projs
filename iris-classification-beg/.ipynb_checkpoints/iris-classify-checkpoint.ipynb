```python
import numpy as np
import pandas as pd
import scipy
import sklearn as skl
import matplotlib.pyplot as plt
from re import sub
import seaborn as sns
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

Following standard practice, i am going to seperate a training, validation, and test set. We have 150 samples, so a split might be 105 in the training set, 25 in the test set, 20 in the validation set. I will use the validation set to test our model for overfitting, and iterate on my design until it seems to have reached a good result, then i will finally test on the test set.

```python
# there is code to do this in the sklearn dataset, im using a random int for reproducability
# also the stratify term, ensures we have an even number of each
train_df, temp_df = skl.model_selection.train_test_split(df, test_size=0.3, random_state=20250320, stratify=df["Species"])
# i want a test set a little larger than the val set
val_df, test_df = skl.model_selection.train_test_split(temp_df, test_size=0.6, random_state=20250320, stratify=temp_df["Species"])
```

To Start, I'm going to plot a pair plot on the training set to visualize all possible relationships, maybe to identify possible points, im dropping id, as i don't particularly want to see how the abitrary id relates to the data

```python
sns.pairplot(train_df.drop(columns=["Id"]), hue="Species")
```

Looking at the plots, the ones that seem to be the most seperable seem to be the petalWidthCm and petalLengthCm, graphs i find promising are (petalLengthCm vs petaWidthCm), (PetalWidthCm vs SepalLengthCm), (petalLengthCm vs sepalLengthCm).

In particular, the seperation of Iris-setosa from the other groups appears simple.

Had there been only 2 things to classify, I would've taken a regression based approach, but it seems that in this case, with 3 different types of flower to separate, I might try a Decision Tree, k-nearest neighbors, SVM, then for fun, i may also implement a Neural Network for the classification.

```python
# Note, i learnt after the fact about a function called LabelEncoder in the sklearn.preprocessing library
def enumerateSpecies(species_list):
    map = {
        "Iris-setosa" : 0,
        "Iris-versicolor" : 1,
        "Iris-virginica" : 2,
    }
    res = []
    for flower in species_list:
        res.append(map[flower])
    return res

# I want to get seperate the features and columsn from all our sets.
feature_columns = [col for col in df.columns if col not in ["Id", "Species"]]
X_train, y_train = train_df[feature_columns], enumerateSpecies(train_df["Species"])
X_val, y_val = val_df[feature_columns], enumerateSpecies(val_df["Species"])
X_test, y_test = test_df[feature_columns], enumerateSpecies(test_df["Species"])
```

```python
dec_Tree = skl.tree.DecisionTreeClassifier()
dec_Tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,10))
skl.tree.plot_tree(dec_Tree, feature_names=train_df.columns, class_names=["Iris-setosa", "Iris-versicolor", "Iris-virginica"], filled=True)
plt.show()
```

```python
y_check = dec_Tree.predict(X_val)
acc = skl.metrics.accuracy_score(y_val, y_check)
print(f"Accuracy on val set: {acc * 100:.3f}%")
```

I'm happy with this, i'll now test it with the test set.

```python
y_pred = dec_Tree.predict(X_test)
acc_test = skl.metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {acc_test * 100:.3f}%")
```

As expected, its slightly worse, but its still solid.

I will now try the k-nearest neighbors method, and see how it compares.

```python
knn = skl.neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_val_knn = knn.predict(X_val)
acc_val_knn = skl.metrics.accuracy_score(y_val, y_val_knn)
print(f"Accuracy on val set: {acc_val_knn * 100:.3f}%")
```

```python
y_test_knn = knn.predict(X_test)
acc_test_knn = skl.metrics.accuracy_score(y_test, y_test_knn)
print(f"Accuracy on test set: {acc_test_knn * 100:.3f}%")
```

```python
# It appears to be a slight improvement, but i would like to see how well we can do with only the PetalWidthCm and PetalLengthCm
selected_features = ["PetalWidthCm", "PetalLengthCm"]
X_train_2d = X_train[selected_features]
X_val_2d = X_val[selected_features]
X_test_2d = X_test[selected_features]

# now the dataset shouldn't have any other dimensions, I have chosen this graph as the groupings of the 3 species appear the most distinct
knn_2d = skl.neighbors.KNeighborsClassifier(n_neighbors=3)
knn_2d.fit(X_train_2d, y_train)

#validation
y_val_pred = knn_2d.predict(X_val_2d)
acc_val_knn2d = skl.metrics.accuracy_score(y_val, y_val_pred)
print(f"Accuracy on val set: {acc_val_knn2d * 100:.3f}%")

#test
y_test_pred = knn_2d.predict(X_test_2d)
acc_test_knn2d = skl.metrics.accuracy_score(y_test, y_test_pred)
print(f"Accuracy on val set: {acc_test_knn2d * 100:.3f}%")
```

For some strange reason I'm getting the same values for accuracy in the 2 models. using KNN 2d and the other one. This could possibly be due to a similar choice in dimension for classification (i.e. the decision tree identified the best split as the same one that i chose?), or simple user error.

Either way, i am happy with the performance with both.

```python
# i am now going to test with a SVM, but a linear kernel will of course not work, so i will use rbf
rbf_svc = skl.svm.SVC()
rbf_svc.fit(X_train, y_train)
y_val_pred = rbf_svc.predict(X_val)
acc_val_svc = skl.metrics.accuracy_score(y_val, y_val_pred)
print(f"Accuracy on val set: {acc_val_svc * 100:.3f}%")

y_test_pred = rbf_svc.predict(X_test)
acc_test_svc = skl.metrics.accuracy_score(y_test, y_test_pred)
print(f"Accuracy on test set: {acc_test_svc * 100:.3f}%")
```

The overall accuracy for the k-nearest neighbors is the highest, I am still getting repeated accuracies though.

For fun, I will now also try implementing a neural network for this dataset. TBH, in practicality, this would be not optimal, we achieve sufficiently high accuracy, I expect the NN to have an accuracy in the 95% range as well, just since some datapoints seems to intersect, and i dont see any meaningful way that a neural network will be able to classify some of these (maybe (hopefully) it will, we will see).

```python
# Set it up
clf = skl.neural_network.MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
# Predicting
y_pred_valnn = clf.predict(X_val)
val_acc_nn = skl.metrics.accuracy_score(y_val, y_pred_valnn)

y_pred_testnn = clf.predict(X_test)
test_acc_nn = skl.metrics.accuracy_score(y_test, y_pred_testnn)
print(f"Validation Acc: {val_acc_nn * 100:.3f}%, Test accuracy: {test_acc_nn * 100:.3f}")
```
